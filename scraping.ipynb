```python

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Cybersecurity rules
URL = "https://infosecawareness.in/cyber-laws-of-india"
r = requests.get(URL)
soup = BeautifulSoup(r.content, 'html.parser') # If this line causes an error, run 'pip install html5lib' or install html5lib
```

```python
#Returns an iterator object
headings = soup.find_all("h4",class_="panel-title")
rules_cyber=soup.find_all("div",class_="panel-body")
headings1=[]
rules_cyber1=[]
for i in headings:
    headings1.append(i.text.strip())
for i in rules_cyber:
    rules_cyber1.append(i.text.strip())
df_cyber=pd.DataFrame(list(zip(headings1,rules_cyber1)), columns=['Headings','Rules'])

```
